```python
import trimesh
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

```python
# Generate pont Clouds
q_points = 10000
center = np.array([0,0,0])
radius = 1
theta = np.random.uniform(0,2*np.pi, q_points)
phi = np.random.uniform(0, np.pi, q_points)
x = radius*np.sin(phi)*np.cos(theta) + center[0]
y = radius*np.sin(phi)*np.sin(theta) + center[1]
z = radius*np.cos(phi) + center[2]
point_cloud = np.column_stack((x,y,z))
```

```python
# render point cloud
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z, c='r', marker='o')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
# Create a mesh from cloud of points
mesh = trimesh.points.plot_points(point_cloud, show=True)
```
